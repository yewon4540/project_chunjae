```python
df = pd.read_parquet('./data/test.parquet')
df = df[:200]
df.head()
```

```python
import time
```

```python
def time_use(nlpy, b):
    start = time.time()
    for i in  range(0, 200): #문서의 형태, 크기에 따라서 다르게 설정하기
        contents=df['contents'].loc[i] #말뭉치 설정
        nouns=nlpy.nouns(contents)
        globals()['Var_{0}{1}'.format(b, i)]=nouns
    end = time.time()
    total_time=(f"{end - start:.5f} sec")
    print('{0} ==> {1}'.format(nlpy ,total_time))
```

```python
# Komoran.nouns('This is test sentence')
```

```python
time_use(okt,'okt')
time_use(hannanum, 'hannanum')
time_use(kkma, 'kkma')
# time_use(komoran, 'komoran')
```
